This lecture solves example problems involving CSTR design for reactions that have non-power law kinetics. Before starting, it may be helpful to review the basic concept of "what is a rate law?" presented in [Notebook 16](https://jqbond.github.io/CEN587/Notebooks/587-N16.html) Sometimes, when we first see complex rate laws, we panic because they don't look like the usual $r = k{C_A}^\alpha$ power law expressions that we now feel pretty good about. Just remember that rate laws are essentially functions that take temperature, pressure, and/or composition as arguments and return the rate of reaction as an output.

## Example Problem 01

The liquid-phase reaction below is carried out in a well-mixed CSTR. 

$$A + 2B \rightarrow C$$

The reaction has non-elementary kinetics, specifically:

$$r = \frac{kC_AC_B}{1+K_1C_A+K_2C_B}$$ 

Data available for this reaction:

\begin{align}
    k &= 7.24 \times 10^{-4} \ \textrm{L} \ \textrm{mol}^{-1} \ \textrm{min}^{-1}\\
    K_1 &= 14.75 \ \textrm{L} \ \textrm{mol}^{-1}\\
    K_2 &= 9.24 \ \textrm{L} \ \textrm{mol}^{-1}\\
    \rho_A &= 0.921 \ \textrm{g} \ \textrm{mL}^{-1}\\
    \rho_B &= 1.234 \ \textrm{g} \ \textrm{mL}^{-1}\\
    \rho_C &= 0.827 \ \textrm{g} \ \textrm{mL}^{-1}\\	
    MW_A &= 97.6 \ \textrm{g} \ \textrm{mol}^{-1}\\
    MW_B &= 84.3 \ \textrm{g} \ \textrm{mol}^{-1}\\
    MW_C &= 266.2 \ \textrm{g} \ \textrm{mol}^{-1}\\
\end{align}

The volumetric flowrate into the reactor is 6 liters per minute, and species concentrations in the influent stream are $C_{Af} = 2.2\mathrm{M}$ and $C_{Bf} = 3.8\mathrm{M}$. What CSTR volume is required to achieve 80\% conversion of species $B$ at steady state? 

**Answer**: $V = 264,000\mathrm{L}$

### Solution to Example Problem 01

Since the problem asks for fractional conversion on $B$, let's try starting the solution by writing a material balance on $B$. For a perfectly mixed CSTR at steady state, we know that:

$$0 = F_{B,f} - F_B + R_BV$$

As we think about this equation, we should recognize that we can calculate the feed molar flowrate of $B$ relatively easily -- $(F_{Bf} = C_{Bf}Q_f)$. The remaining terms, $F_B$, $R_B$, and $V$ are unknown right now. To solve for their values, we need to either add more equations that we can solve simultaneously with this one, or we can write expressions that allow us to reduce the number of unknowns to 1. Let's see if we can figure out a strategy for doing this here. First, we can define $R_B$ as $R_B = \sum_i \nu_{i,B}r_i$ to find:

$$R_B = -2r$$

The rate law is complicated, but it is still just a rate law. We pass concentrations to it, and it tells us the rate in units of moles per volume per time.

$$r = \frac{kC_AC_B}{1+K_1C_A+K_2C_B}$$

To use it, we need to define concentrations.  We'll do so using the exit flowrates from the CSTR:

\begin{align}
    C_A = F_A/Q \\
    C_B = F_B/Q \\
\end{align}

Here, we're not so lucky to have a constant density fluid, so we need to calculate $Q$ at the reactor exti. We'll do so by assuming an ideal solution, which effectively means that the volumes occupied by each species can be added to calculate the total volume of the liquid. Analogously, if we know the volumetric flowrates of each species, $Q_j$, we sum them up to get the total volumetric flowrate, $Q$.

$$Q = \sum_j Q_j$$

We can express the volumetric flowrate of each species as:

$$Q_j = F_j{V_j}^\circ$$

Substitution into the summation gives:

$$Q = \sum_j F_j{V_j}^\circ$$

Which, for our system expands to:

$$Q = F_A{V_A}^\circ + F_B{V_B}^\circ + F_C{V_C}^\circ$$

We can define the molar volume of any species from its density, $\rho_j$, and its molecular weight, $MW_j$:

$$V_j = \frac{MW_j}{\rho_j}$$

Now we see that $Q$ depends on the molar flowrates of $A$, $B$, and $C$. That means our defintion of the concentrations of $A$ and $B$ depend on the molar flowrates of all species.  So if we look back at our original material balance:

$$0 = F_{B,f} - F_B + R_BV$$

It has two obvious unknowns: $F_B$ and $V$. It also has two unknowns buried in the definition of $R_B$: These are $F_B$ and $F_C$. If we take stock of things now, it looks like we have 4 unknowns and only one equation. We can either write more equations, or we can reduce the number of unknowns.  Let's chose the second approach here because it is an algebraic equation, and it is usually easer to deal with the smallest set of unknowns possible. An important additional observation is that the exit fractional conversion is specified in this problem ($X_B = 0.8$). Since this process involves only one reaction, this specification means that we can solve for the exit flowrates of all species based on the conversion of $B$ attained in the reactor. This will simplify our solution considerably!!

We express all molar flowrates as a function of fractional conversion of $B$, $X_B$ and organize the information in a mole table. The resulting expressions are:

\begin{align}
    F_A &= F_{A,f} - 1/2F_{B,f}X_B \\
    F_B &= F_{B,f} - F_{B,f}X_B \\
    F_C &= F_{C,f} + 1/2F_{B,f}X_B \\
\end{align}

We calculate feed molar flowrates for all species based on information in the problem statement:

$$F_{j,f} = C_{j,f}Q_f$$

And we know the problems is asking for a conversion of 80%. As mentioned above, we can calculate the exit molar flowrate of each species with almost no effort using results from the mole table. That leaves us with only one unknown in the material balance, $V$, so we solve for it explicitly:

$$V = \frac{-(F_{B,f} - F_B)}{R_B}$$

If you examine this equation carefully, you'll see that once we substitute $X_B = 0.8$ into the mole table expressions, the numerical value of each term on the right hand side of the equation can be calculated, so the problem is effectively solved.

````python
#########################################################################
# This cell defines/calculates a set of global variables/constants      #
#########################################################################

#Feed Volumetric Flowrate
Qf  = 6.0   #L/min

#Feed Concentrations
CAf = 2.2   #mol/L
CBf = 3.8   #mol/L
CCf = 0.0   #mol/L

#Feed Molar Flowrates
FAf = CAf*Qf
FBf = CBf*Qf
FCf = CCf*Qf

#Kinetic/thermodynamic parameters
k  = 7.24e-4 #L/mol/min
K1 = 14.75  #L/mol
K2 = 9.24   #L/mol

#Densities
ρA = 0.921*1000   #g/L
ρB = 1.234*1000  #g/L
ρC = 0.827*1000   #g/L

#Molecular Weights
MWA  = 97.6  #g/mol
MWB  = 84.3  #g/mol 
MWC  = 266.2 #g/mol

#Calculating molar volumes
VA = MWA/ρA #L/mol
VB = MWB/ρB #L/mol
VC = MWC/ρC #L/mol

##############################################################################
# I've chosen to write a function that calculates volume once given XB       #
##############################################################################

def Volume_calc_1(XB):
    
    #Exit molar flowrates
    FA = FAf - 1/2*FBf*XB
    FB = FBf - FBf*XB
    FC = FCf + 1/2*FBf*XB

    #Exit Volumetric Flowrate
    Q = FA*VA + FB*VB + FC*VC

    #Exit Concentrations
    CA = FA/Q
    CB = FB/Q

    #Rate of reaction
    r = k*CA*CB/(1 + K1*CA + K2*CB)

    #Production rate of A
    RB = -2*r

    #Solve for Volume
    V = (FBf - FB)/-RB
    return V

XB1 = 0.8
V1a = Volume_calc_1(XB1)

print(f'To achieve XB = {XB1:0.2f} at steady state, the CSTR should have a volume of {V1a:0.2E}L') 
````

One nice thing about having an analytical solution is that it makes it very easy to extend our analysis to other situations where we may require a different fractional conversion. For example, having the function defined above, we can immediately calculate the CSTR volume in Liters for any desired conversion of B!

### Alternate Numerical Solution to Example Problem 01

```{tip}
We usually have to make a choice about how we will solve any given problem. Often, multiple options are available. Should we solve it analytically or numerically? If we solve it numerically, how should we construct the objective function? What will the unknowns be?  Here, we are presenting an alternate solution to the above problem using a nonlinear equation solver. This is not mandatory -- it is a different way to solve the problem, and it is included here to help us see that any problem solution can potentially take many correct trajectories. The one that we choose usually comes down to preference or convenience, but we should start to develop a sense that any of them can be correct as long as we see things through.
```

As discussed in the first solution, another strategy would be to develop a system of equations that could be solved simultaneously. There is nothing wrong with this approach; we can certainly solve the problem this way, and we will do that now. As we work through this solution, consider the differences in the two approaches and the situations where one might be more straightforward than another. In practice, we may sometimes find that it is *slightly* more challenging to add equations in a system of algebraic equations, but it is usually something we can handle. Let's go ahead and develop the system of equations for numerical solution with `opt.root()` and consider the advantages and disadvantages to the approach as we do so.

We start by writing balances on each species:

\begin{align}
    0 = F_{A,f} - F_A + R_AV \\
    0 = F_{B,f} - F_B + R_BV \\
    0 = F_{C,f} - F_C + R_CV \\
\end{align}

This gives us 3 equations written in terms of 4 obvious unknowns $(F_A, F_B, F_C, V)$. We also should recognize that production rates, $R_j$ are also unknown, but at this point, we should generally recognize that, in a CSTR at steady state, these can be expressed as function of molar and volumetric flowrates, so they are not really additional unknowns per se. We'll add one more equation, which is the process specification that we have to achieve 80% conversion of $B$.

\begin{align}
    0 &= F_{A,f} - F_A + R_AV \\
    0 &= F_{B,f} - F_B + R_BV \\
    0 &= F_{C,f} - F_C + R_CV \\
    0 &= X_B - 0.8
\end{align}

While it looks like there are more than 4 unknowns here, we can define everything on the right hand side of these equations in terms of $F_A$, $F_B$, $F_C$, and $V$. Hence, we have now written 4 equations in terms of 4 unknowns, and the problem can be solved. The rest of the process is expressing everything on the right hand side of those equations in terms of those 4 unknowns. First, we define production rates are defined in terms of reaction rates and stoichiometry:

\begin{align}
    R_A &= -r  \\
    R_B &= -2r \\
    R_C &= r   \\
\end{align}

We are given a rate law that defines $r = f(C_A, C_B)$:

$$r = \frac{kC_AC_B}{1+K_1C_A+K_2C_B}$$

To use it, we need to define concentrations.  We'll do so using the exit flowrates from the CSTR:

\begin{align}
    C_A = F_A/Q \\
    C_B = F_B/Q \\
\end{align}

We calculate Q by assuming this is a perfect mixture. Since we are given densities and molecular weights, we can calculate molar volumes and then use these to calculate a total volumetric flowrate:

$$V_j = \frac{MW_j}{\rho_j}$$

And then:

$$Q = F_AV_A + F_BV_B + F_CV_C$$

Finally, we define fractional conversion of B in terms of flowrates:

$$X_B = \frac{F_{B,f} - F_B}{F_{B,f}}$$

That's basically it! Now we'll set up a system of 4 equations and 4 unknowns and solve everything simultaneously with `opt.root()`.

````python
###############################################################################
# Define an objective function that encodes our system of equations           #
# This is of the form f([FA, FB, FC, V])                                      #
# It returns the LHS of the 4 algebraic equations taht we're trying to solve  #
###############################################################################

def P01(var):
    FA, FB, FC, V = var
    XB_target = 0.8

    #Conversion of B
    XB = (FBf - FB)/FBf

    #Exit Volumetric Flowrate
    Q = FA*VA + FB*VB + FC*VC

    #Exit Concentrations
    CA = FA/Q
    CB = FB/Q

    #Rate of reaction
    r = k*CA*CB/(1 + K1*CA + K2*CB)

    #Production rates
    RA = -1*r
    RB = -2*r
    RC =  1*r

    #System of Equations
    LHS1 = FAf - FA + RA*V
    LHS2 = FBf - FB + RB*V
    LHS3 = FCf - FC + RC*V
    LHS4 = XB  - XB_target
    return [LHS1, LHS2, LHS3, LHS4]


#############################################################################################################
# One of the most important challenges with adding more equations is the need for initial guesses!          #
# Generally, for nonlinear algebraic equations, initial guesses must be good-to-great.                      #
# This is difficult for us to accomplish as we may not have a good idea of the correct answer at the outset #
#############################################################################################################

var0  = [FAf - FBf*0.8/2, FBf*0.2, FBf*0.8/2, 10]

###################################################################################
# Solve the problem using opt.root()                                              #
###################################################################################

ans1b = opt.root(P01, var0)
V1b   = ans1b.x[-1] #Volume is the final unknown in the .x "answer" field

print(ans1b, '\n')
print(f'To achieve XB = {XB1:0.2f} at steady state, the CSTR should have a volume of {V1b:0.2E}L') 
````

### Some advantages of the analytical solution and/or challenges with the numerical solution

Imagine that instead of answering the question of what CSTR volume is required to accomplish $X_B = 0.8$, we were asked to plot the relationship between CSTR volume and fractional conversion for a range of fractional conversions $X_B = [0.0, 0.9]$. This would be a very useful piece of information as it would allow us to quickly assess how changing our conversion specification would impact the size of reactor required to accomplish that specification.

Let's compare the case where we've solved the problem analytically to obtain $V(X_B)$ to the one that we solved numerically!

#### Analytical Solution

As shown above, we were able to develop a function:

$$V(X_B) = \frac{-(F_{B,f} - F_B)}{R_B}$$

The beauty of this function is that once we pass a specific value of $X_B$ as an argument, all terms of the right hand side can be calculated numerically. It is then straightforward to broadcast this function to the range of conversions $X_B = [0.0, 0.9]$.

````python
############################################################################
# Graphing the CSTR volume required to achieve XB = 0 to 0.9               #
############################################################################

xbset1 = np.linspace(0, 0.90, 100)
plt.plot(xbset1, Volume_calc_1(xbset1), color = 'black', ls = 'solid', label = 'analytical')
plt.xlim(0, 1.0)
plt.ylim(0, 600000)
plt.xlabel('Fractional Conversion of B', fontsize = 14)
plt.ylabel('CSTR Volume (L)', fontsize = 14)
plt.legend(loc = 'upper left')
plt.show()
````

#### Numerical Solution

Now lets say that we are not able to solve the problem analytically, and we will try to solve it numerically for the set of conversions $X_B = [0.0, 0.9]$.. Conceptually, we are still approaching the problem the same way -- we are solving algebraic equations for a given value of conversion. But in this case, doing so means we need to use `opt.root()` to solve the system of equations at each conversion in $X_B = [0.0, 0.9]$. There are multiple ways we can do this, but the most conceptually straightforward is to write a for loop to solve a new instance of the system of equations for each value of $X_B \ \mathrm{in} \ [0.0, 0.9]$.  We'll do this by constructing our objective function such that it takes "extra parameters." We'll then use the `args` keyword to pass a new value of fractional conversion of B on each pass through the loop. In addition, we'll store the answer for $V$ on each pass through the loop.  This will generate a set of $V$ for each value of $X_B$, and we can overlay that discrete set of $(X_B, V)$ pairs with the analytical solution, which is a continuous function $V(X_B)$.

````python
###########################################################################################
# modified system of equations function for use with opt_root() and args keyword          #
# It is of the form f([FA, FB, FC, V], extra_arguments)                                   #
# It still returns the LHS of each equation in the system                                 #
###########################################################################################

def P01b(var, XB_target):
    FA, FB, FC, V = var

    #Conversion of B
    XB = (FBf - FB)/FBf

    #Exit Volumetric Flowrate
    Q = FA*VA + FB*VB + FC*VC

    #Exit Concentrations
    CA = FA/Q
    CB = FB/Q

    #Rate of reaction
    r = k*CA*CB/(1 + K1*CA + K2*CB)

    #Production rates
    RA = -1*r
    RB = -2*r
    RC =  1*r

    #System of Equations
    LHS1 = FAf - FA + RA*V
    LHS2 = FBf - FB + RB*V
    LHS3 = FCf - FC + RC*V
    LHS4 = XB  - XB_target
    return [LHS1, LHS2, LHS3, LHS4]


#################################################################################
# Define a discrete set of fractional conversions for which to solve the system #
#################################################################################

xbset2 = np.linspace(0, 0.9, 20)

#################################################################################
# Initialize a numpy array where the answer (V) can be stored on each pass      #
#################################################################################

V1out = np.zeros(len(xbset2))

#################################################################################
# Loop over all values in xbset; solve the system of equations at each xb       #
# Store the volume answer in V1out                                              #
#################################################################################

for i in range(0, len(xbset2), 1):
    xb       = xbset2[i]
    var0     = [FAf - FBf*xb/2, FBf - FBf*xb, FBf*xb, 10]
    ans      = opt.root(P01b, var0, args = (xb,))
    V1out[i] = ans.x[-1]

#################################################################################
# Overlay numerical and analytical solutions                                    #
#################################################################################

plt.plot(xbset1, Volume_calc_1(xbset1), color = 'black', ls = 'solid', label = 'analytical')
plt.scatter(xbset2, V1out, color = 'red', marker = 'o', label = 'opt.root()')
plt.xlim(0, 1.0)
plt.ylim(0, 600000)
plt.xlabel('Fractional Conversion of B', fontsize = 14)
plt.ylabel('CSTR Volume (L)', fontsize = 14)
plt.legend(loc = 'upper left')
plt.show()
````

## Example Problem 02

The same liquid-phase reaction from above is carried out in a well-mixed CSTR. 

$$A + 2B \rightarrow C$$

The reaction has non-elementary kinetics, specifically:

$$r = \frac{kC_AC_B}{1+K_1C_A+K_2C_B}$$ 

Data available for this reaction:

\begin{align}
    k &= 7.24 \times 10^{-4} \ \textrm{L} \ \textrm{mol}^{-1} \ \textrm{min}^{-1}\\
    K_1 &= 14.75 \ \textrm{L} \ \textrm{mol}^{-1}\\
    K_2 &= 9.24 \ \textrm{L} \ \textrm{mol}^{-1}\\
    \rho_A &= 0.921 \ \textrm{g} \ \textrm{mL}^{-1}\\
    \rho_B &= 1.234 \ \textrm{g} \ \textrm{mL}^{-1}\\
    \rho_C &= 0.827 \ \textrm{g} \ \textrm{mL}^{-1}\\	
    MW_A &= 97.6 \ \textrm{g} \ \textrm{mol}^{-1}\\
    MW_B &= 84.3 \ \textrm{g} \ \textrm{mol}^{-1}\\
    MW_C &= 266.2 \ \textrm{g} \ \textrm{mol}^{-1}\\
\end{align}

The volumetric flowrate into the reactor is 6 liters per minute, and species concentrations in the influent stream are $C_{Af} = 2.2\mathrm{M}$ and $C_{Bf} = 3.8\mathrm{M}$. What is the fractional conversion of $B$ achieved at steady state in a $164 \ \mathrm{m^3}$ CSTR? 

**Answer**: $X_B = 0.706$

### Solution to Example Problem 02

The basic stragety here is similar to that in [Example Problem 01](#example-problem-01). We will notice that actually solving the equations is a bit harder in this case. Regardless, we start the problem the same way, by writing a material balance on $B$ in a perfectly mixed CSTR at steady state:

$$0 = F_{B,f} - F_B + R_BV$$

We define $R_B$ to find:

$$R_B = -2r$$

The rate law is given below.

$$r = \frac{kC_AC_B}{1+K_1C_A+K_2C_B}$$

We define concentrations using the exit flowrates from the CSTR:

\begin{align}
    C_A = F_A/Q \\
    C_B = F_B/Q \\
\end{align}

Again, since density is not constant, we apply an equation of state to get the volumetric flowrate. Here, we use the ideal mixing assumption (volumes are additive).

$$Q = \sum_j Q_j$$

We can express the volumetric flowrate of each species as:

$$Q_j = F_j{V_j}^\circ$$

Substitution into the summation gives:

$$Q = \sum_j F_j{V_j}^\circ$$

Which, for our system expands to:

$$Q = F_A{V_A}^\circ + F_B{V_B}^\circ + F_C{V_C}^\circ$$

We note that we can define the molar volume of any species from its density, $\rho_j$, and its molecular weight, $MW_j$:

$$V_j = \frac{MW_j}{\rho_j}$$

$Q$ depends on the molar flowrates of $A$, $B$, and $C$.  That means our defintion of the concentrations of $A$ and $B$ depend on all species, etc.  So if we look back at our original material balance:

$$0 = F_{B,f} - F_B + R_BV$$

It has one obvious unknown: $F_B$.  It also has two unknowns buried in the definition of $R_B$: These are $F_A$ and $F_C$.  So, for this problem we have 3 unknowns and only one equation.  We can either write more equations, or we can reduce the number of unknowns. Let's choose the second approach here. A good reason to reduce the number of unknowns is that this is an algebraic system, and it is usually advantageous to the smallest set of unknowns possible. We can reduce the problem to a single unknown by expressing all molar flowrates as a function of fractional conversion of $B$, $X_B$, much as we did in [Example Problem 01](#example-problem-01). The resultant mole table is:

\begin{align}
    F_A &= F_{A,f} - 1/2F_{B,f}X_B \\
    F_B &= F_{B,f} - F_{B,f}X_B \\
    F_C &= F_{C,f} + 1/2F_{B,f}X_B \\
\end{align}

Here we know the volume of the reactor is 164,000L. Unfortunately, the problem is difficult to solve by hand despite having a single unknown. This is because it involves back substitution of numerous functions of conversion into the rate expression and material balance. For this reason, we will solve numerically despite having a univariate problem that can probably (possibly?) be solved analytically. It is generally preferable to write a readable function and allow Python to make substitutions instead of doing a hand substitutions on paper and developing a physically non-intuitive equation to solve using `opt.newton()`:

````python
########################################################################################
# Developing univariate objective function for use with opt.newton()                   #
# It is of the form f(XB), and it returns LHS = FBf - FB + RB*V                        #
# opt.newton() will find the value of XB where 0 = FBf - FB + RB*V                     #
########################################################################################

def P02a(XB):
    
    #Reactor Volume
    V   = 164*1000 #L

    #Exit molar flowrates
    FA = FAf - 1/2*FBf*XB
    FB = FBf - FBf*XB
    FC = FCf + 1/2*FBf*XB
    
    #Exit Volumetric Flowrate
    Q = FA*VA + FB*VB + FC*VC

    #Exit Concentrations
    CA = FA/Q
    CB = FB/Q

    #Rate of reaction
    r = k*CA*CB/(1 + K1*CA + K2*CB)

    #Production rates
    RA = -1*r
    RB = -2*r
    RC =  1*r

    #Equation to solve is a material balance on B, i.e., 0 = FBf - FB + RB*V:
    LHS = FBf - FB + RB*V
    return LHS

V2   = 164*1000 #L
XB0  = 0.65
ans2a, info = opt.newton(P02a, XB0, full_output = True)
print(info, '\n')
print(f'A {V2}L CSTR will achieve a fractional conversion of {ans2a:0.3f} at steady state')
````

### An Alternate Solution to Example 02

```{tip}

```

We will again start the problem  by writing a material balance on $B$. For a perfectly mixed CSTR at steady state:

$$Q = \sum_j Q_j$$

We can express the volumetric flowrate of each species as:

$$Q_j = F_j{V_j}^\circ$$

Substitution into the summation gives:

$$Q = \sum_j F_j{V_j}^\circ$$

Which, for our system expands to:

$$Q = F_A{V_A}^\circ + F_B{V_B}^\circ + F_C{V_C}^\circ$$

We note that we can define the molar volume of any species from its density, $\rho_j$, and its molecular weight, $MW_j$:

$$V_j = \frac{MW_j}{\rho_j}$$

$Q$ depends on the molar flowrates of $A$, $B$, and $C$.  That means our defintion of the concentrations of $A$ and $B$ depend on all species, etc.  So if we look back at our original material balance:

$$0 = F_{B,f} - F_B + R_BV$$

It has one obvious unknown: $F_B$. It also has two unknowns buried in the definition of $R_B$: These are $F_A$ and $F_C$. We are back at the point of having 3 unknowns and only one equation. Above, we resolved this by expressing all unknown molar flowrates as functions of a single unknown fractional conversion, $X_B$. Here, we will instead write additional material balances on $A$ and $C$, which will give us 3 equations expressed in terms of 3 unknowns:

\begin{align}
    0 = F_{A,f} - F_A + R_AV \\
    0 = F_{B,f} - F_B + R_BV \\
    0 = F_{C,f} - F_C + R_CV \\
\end{align}

We know the volume of the reactor is 164,000L.  All terms in the above three equations are either constants (feed molar flowrates, $V$, rate constants, equilibrium constants, molar volumes, etc.) or they are our three unknowns $(F_A, F_B, F_C)$. This is therefore a system of 3 equations and 3 unknowns, so we can solve with `opt.root()`

````python
###################################################################################################
# Develop an objective function that defines our system of equations to be solved with opt.root() #
# This function has the form f([FA, FB, FC]), and it returns the LHS of each material balance     #
###################################################################################################

def P02b(var):
    FA, FB, FC = var

    #Reactor Volume
    V   = 164*1000 #L
    
    #Exit Volumetric Flowrate
    Q = FA*VA + FB*VB + FC*VC

    #Exit Concentrations
    CA = FA/Q
    CB = FB/Q

    #Rate of reaction
    r = k*CA*CB/(1 + K1*CA + K2*CB)

    #Production rates
    RA = -1*r
    RB = -2*r
    RC =  1*r

    #System of Equations
    LHS1 = FAf - FA + RA*V
    LHS2 = FBf - FB + RB*V
    LHS3 = FCf - FC + RC*V
    return [LHS1, LHS2, LHS3]

#############################################################################
# Generate reasonable initial guesses by assuming an exit conversion of 0.6 #
#############################################################################
XB0   = 0.6
var0  = [FAf - FBf*XB0, FBf*(1 - XB0), FBf*XB0]

#############################################################################
# Solve using opt.root()                                                    #
#############################################################################

ans2b = opt.root(P02b, var0)
print(ans2b, '\n')
FB2b  = ans2b.x[1]
XB2b  = (FBf - FB2b)/FBf 
print(f'A {V2}L CSTR will achieve a fractional conversion of {XB2b:0.3f} at steady state')
````
